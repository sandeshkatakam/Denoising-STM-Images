```python
from ipywidgets import interact, widgets
from IPython.display import display
def visualize_stm_img(image_no):
    sample_df = pd.read_csv('C:/Users/sande/Downloads/MyGitHub/Denoising-STM-Images/data/10818A01_G_0sub_ft_amp_ctr_rm_'+str(image_no)+'.csv')
    arr = sample_df.to_numpy()
    plt.figure(figsize = (10,13))
    plt.axis('off')
    plt.title('Image Layer Number')
    plt.imshow(arr, cmap = 'YlOrRd')

interact(visualize_stm_img, image_no=widgets.IntSlider(min=1,max=11,step=1,value=1));
```
